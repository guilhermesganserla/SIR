```python
# packages
import scipy.integrate
import numpy
import matplotlib.pyplot as plt
```

```python
# ODEs
def SIR_model (y, t, beta, gamma):
    S, I, R = y
    dS_dt = -beta*S*I
    dI_dt = beta*S*I - gamma*I
    dR_dt = gamma*I
    
    return([dS_dt, dI_dt, dR_dt])   

# Onde S são os suscetíveis, I os infectados e R os recuperados, somando-se tem a população total.
# Beta é a taxa de transmissão e gamma a de recuperação.
```

```python
# Condições iniciais
S0 = 0.99
I0 = 0.01
R0 = 0
beta = 0.325
gamma = 0.1

# Vetor tempo
t = numpy.linspace(0, 100, 10000)

#Resultado

solution = scipy.integrate.odeint(SIR_model, [S0,I0,R0], t, args=(beta, gamma))
solution = numpy.array(solution)
```

```python
# Gráfico
plt.figure(figsize=[6,4])
plt.plot(t,solution[:,0], label="S(t)")
plt.plot(t,solution[:,1], label="I(t)")
plt.plot(t,solution[:,2], label="R(t)")
plt.grid()
plt.legend()
plt.xlabel("Dias")
plt.ylabel("Proporção da população infectada")

plt.show()
```
